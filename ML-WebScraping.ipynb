```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import csv

page=requests.get("https://www.espncricinfo.com/player/ms-dhoni-28081/matches")
output_file=csv.writer(open('MSDhoni.csv','w+'))

output_file.writerow(['Match','Bat','Wicket','Date','ground','Format'])
soup=BeautifulSoup(page.text,'html.parser')
table=soup.find_all("table")

league_table=table[0]

teams=league_table.find_all("tr")
teams

for team in teams[1:100]:
    stats=team.find_all('td')
    Match=stats[0].text
    Bat=stats[1].text 
    Wicket=stats[2].text
    Date=stats[3].text
    ground=stats[4].text     
    Format=stats[5].text 
    
    print(Match,Bat,Wicket,Date,ground,Format)
    output_file.writerow([Match,Bat,Wicket,Date,ground,Format])
```

```python
df = pd.read_csv("MSDhoni.csv")
```

```python
df
```
