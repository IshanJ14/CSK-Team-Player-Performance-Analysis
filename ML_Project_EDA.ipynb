```python
import pandas as pd
import re
import plotly.express as px
import plotly.io as pio
pio.renderers.default='browser'

def clean(df,name):
    Val = []
    for value in df['Bat']:
        if str(value) == '--':
            Val.append(0)
        else:
            temp = re.findall(r'\d+', str(value))
            Val.append(int(temp[0]))
    return pd.DataFrame(list(zip(df['Date'], Val)),columns =['Date', name])

df1 = clean(pd.read_csv(r"C:\Users\Ishan\Desktop\Python\MSDhoni.csv"),'MSDhoni')
df2 = clean(pd.read_csv(r"C:\Users\Ishan\Desktop\Python\ChrisJordan.csv"),'ChrisJordan')
df3 = clean(pd.read_csv(r"C:\Users\Ishan\Desktop\Python\ShivamDube.csv"),'ShivamDube')
df4 = clean(pd.read_csv(r"C:\Users\Ishan\Desktop\Python\DwainePretorius.csv"),'DwainePretorius')
df5 = clean(pd.read_csv(r"C:\Users\Ishan\Desktop\Python\DwayneBravo.csv"),'DwayneBravo')
df6 = clean(pd.read_csv(r"C:\Users\Ishan\Desktop\Python\MoeenAli.csv"),'MoeenAli')
df7 = clean(pd.read_csv(r"C:\Users\Ishan\Desktop\Python\RavindraJadeja.csv"),'RavindraJadeja')
df8 = clean(pd.read_csv(r"C:\Users\Ishan\Desktop\Python\RobinUthappa.csv"),'RobinUthappa')
df9 = clean(pd.read_csv(r"C:\Users\Ishan\Desktop\Python\RuturajGaikwad.csv"),'RuturajGaikwad')
df10 = clean(pd.read_csv(r"C:\Users\Ishan\Desktop\Python\AmbatiRayudu.csv"),'AmbatiRayudu')

df1.drop([6, 7, 8, 9], inplace = True)
df2.drop([4, 5, 6, 7, 8, 9], inplace = True)
df3.drop([6, 7, 8, 9], inplace = True)
df4.drop([2, 3, 4, 5, 6, 7, 8, 9], inplace = True)
df5.drop([6, 7, 8, 9], inplace = True)
df6.drop([5, 6, 7, 8, 9], inplace = True)
df7.drop([6, 7, 8, 9], inplace = True)
df8.drop([6, 7, 8, 9], inplace = True)
df9.drop([6, 7, 8, 9], inplace = True)
df10.drop([6, 7, 8, 9], inplace = True)

df1 = df1.merge(df2, how='left', on='Date')
df1 = df1.merge(df3, how='left', on='Date')
df1 = df1.merge(df4, how='left', on='Date')
df1 = df1.merge(df5, how='left', on='Date')
df1 = df1.merge(df6, how='left', on='Date')
df1 = df1.merge(df7, how='left', on='Date')
df1 = df1.merge(df8, how='left', on='Date')
df1 = df1.merge(df9, how='left', on='Date')
df1 = df1.merge(df10, how='left', on='Date')

FinalDf= df1
FinalDf = FinalDf.fillna(0)
print(FinalDf)

fig = px.line(FinalDf, x='Date', y=FinalDf.columns[1:11], markers=True)
fig.show()

Y = []
for i in range(1,11):
    Y.append(sum(FinalDf.iloc[:,i]))


fig = px.bar(x=FinalDf.columns[1:11], y=Y, title="Total Runs made by all Players",labels={
                     "x": "Player",
                     "y": "Total Runs"
                 })
fig.show()

fig = px.pie(values=Y, names=FinalDf.columns[1:11], title='Total Percentage of Runs by Each player')
fig.show()
```
